```python
# Convert "Date" to datetime type
data['Date'] = pd.to_datetime(data['Date'])

# Convert categorical columns "Venue" and "Opponent" to unique numeric values and create new columns
data['Venue_Numeric'] = data['Venue'].astype('category').cat.codes
data['Opponent_Numeric'] = data['Opponent'].astype('category').cat.codes

# Extract only the hour from the "Time" column, convert to integer, and create a new column
data['Time_Hour'] = data['Time'].apply(lambda x: int(x.split(':')[0]))

# Convert "Day" to a unique day of the week as numeric and create a new column
days_mapping = {'Mon': 0, 'Tue': 1, 'Wed': 2, 'Thu': 3, 'Fri': 4, 'Sat': 5, 'Sun': 6}
data['Day_Numeric'] = data['Day'].map(days_mapping)

# Map "Result" to the specified values and create a new column
result_mapping = {'W': 2, 'D': 1, 'L': 0}
data['Result_Numeric'] = data['Result'].map(result_mapping)

# Display the updated data with the new columns
data.head()
```

```python
# Defining the features and target variable
features = ['Venue_Numeric', 'Opponent_Numeric', 'Time_Hour', 'Day_Numeric'] #'GF',	'GA', 'Sh',	'SoT', 'Dist', 'FK', 'PK',	'PKatt',
target = 'Result_Numeric'

X = data[features]
y = data[target]

# Splitting the data into training and test sets based on the specified date
train_data = data[data['Date'] <= '2022-01-01']
test_data = data[data['Date'] > '2022-01-01']

X_train_time_based = train_data[features]
y_train_time_based = train_data[target]
X_test_time_based = test_data[features]
y_test_time_based = test_data[target]

```

```python
# Checking the shape of the training and testing sets
X_train_time_based.shape, X_test_time_based.shape, y_train_time_based.shape, y_test_time_based.shape
```

```python
# Checking for missing values in the updated training and test sets
missing_values_train = X_train_time_based.isnull().sum()
missing_values_test = X_test_time_based.isnull().sum()

missing_values_train[missing_values_train > 0] , missing_values_test[missing_values_test > 0]

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

# Defining the parameter grid for hyperparameter tuning
param_grid = {
    'n_estimators': [50, 100, 150],
    'min_samples_split': [2, 5, 10],
    'random_state': [42]
}

# Creating the Random Forest Classifier
clf = RandomForestClassifier()

# Performing grid search with cross-validation
grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5)
grid_search.fit(X_train_time_based, y_train_time_based)

# Retrieving the best hyperparameters
best_params = grid_search.best_params_
best_params
```

```python
# Training the Random Forest model with the optimal hyperparameters
best_rf_model = RandomForestClassifier(
    n_estimators=best_params['n_estimators'],
    min_samples_split=best_params['min_samples_split'],
    random_state=best_params['random_state']
)

best_rf_model.fit(X_train_time_based, y_train_time_based)

# Checking the training accuracy
train_accuracy = best_rf_model.score(X_train_time_based, y_train_time_based)
train_accuracy
```

```python
# Evaluating the model on the test set
test_accuracy = best_rf_model.score(X_test_time_based, y_test_time_based)
test_accuracy
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

# Predicting the results on the test set
y_pred_time_based = best_rf_model.predict(X_test_time_based)


# Calculating the accuracy score
accuracy_time_based = accuracy_score(y_test_time_based, y_pred_time_based)

# Calculating the confusion matrix
confusion_mat_time_based = confusion_matrix(y_test_time_based, y_pred_time_based)

accuracy_time_based, confusion_mat_time_based
```

```python
# Creating a DataFrame for the confusion matrix with proper labels
confusion_matrix_df = pd.DataFrame(
    confusion_mat_time_based,
    index=["Actual Loss","Actual Draw", "Actual Win"],
    columns=["Predicted Loss","Predicted Draw", "Predicted Win"]
)

# Displaying the confusion matrix as a table
confusion_matrix_df
```
