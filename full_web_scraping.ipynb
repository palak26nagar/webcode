```python
import requests
import pandas as pd
import time
from bs4 import BeautifulSoup

def get_soup(url):
    data = requests.get(url)
    return BeautifulSoup(data.text, 'html.parser')

def get_team_urls(standings_soup):
    standings_table = standings_soup.select('table.stats_table')[0]
    links = [l.get("href") for l in standings_table.find_all('a')]
    links = [l for l in links if '/squads/' in l]
    return [f"https://fbref.com{l}" for l in links]

def get_previous_season_url(standings_soup):
    previous_season = standings_soup.select("a.prev")[0].get("href")
    return f"https://fbref.com{previous_season}"

def get_team_data(team_url, year):
    team_name = team_url.split("/")[-1].replace("-Stats", "").replace("-", " ")
    data = requests.get(team_url)
    matches = pd.read_html(data.text, match="Scores & Fixtures")[0]
    matches = matches[matches["Comp"] == "Premier League"]
    matches["Season"] = year
    matches["Team"] = team_name
    return matches

def scrape_premier_league_data(years):
    all_matches = []
    standings_url = "https://fbref.com/en/comps/9/2022-2023/2022-2023-Premier-League-Stats"

    for year in years:
        standings_soup = get_soup(standings_url)
        team_urls = get_team_urls(standings_soup)
        standings_url = get_previous_season_url(standings_soup)

        for team_url in team_urls:
            team_data = get_team_data(team_url, year)
            if team_data is not None:
                all_matches.append(team_data)
            time.sleep(3)

    return pd.concat(all_matches)

years = list(range(2022, 2018, -1))
all_matches_df = scrape_premier_league_data(years)
```

```python
all_matches_df.reset_index(drop=True, inplace=True)  # Reset the index to provide automatic numbering
all_matches_df.to_csv('premier_league.csv')  # Save to CSV

```
