```python
# Function to categorize time into morning, afternoon, evening, night
def categorize_time(time_str):
    time = int(time_str.split(':')[0])
    if 5 <= time < 12:
        return 'Morning'
    elif 12 <= time < 17:
        return 'Afternoon'
    elif 17 <= time < 21:
        return 'Evening'
    else:
        return 'Night'

# Applying the categorization function to the Time column
data['Time_Category'] = data['Time'].apply(categorize_time)

# Checking the first few rows to see the new Time_Category column
data[['Time', 'Time_Category']].head()
```

```python
# Performing one-hot encoding on the Time_Category column
time_categories_one_hot = pd.get_dummies(data['Time_Category'], prefix='Time')

# Concatenating the one-hot encoded columns with the original data
data_one_hot = pd.concat([data, time_categories_one_hot], axis=1)

# Dropping the original Time and Time_Category columns as they are no longer needed
#data_one_hot.drop(columns=['Time', 'Time_Category'], inplace=True)

# Displaying the first few rows to see the one-hot encoded columns
data_one_hot.head()
```

```python
# Encoding the Venue column as 0 for Away and 1 for Home
data_one_hot['Venue_Encoded'] = data_one_hot['Venue'].apply(lambda x: 0 if x == 'Away' else 1)

# Displaying the first few rows to see the original Venue column and the new Venue_Encoded column
data_one_hot[['Venue', 'Venue_Encoded']].head()
```

```python
# Performing one-hot encoding on the Day and Opponent columns
day_one_hot = pd.get_dummies(data_one_hot['Day'], prefix='Day')
opponent_one_hot = pd.get_dummies(data_one_hot['Opponent'], prefix='Opponent')

# Concatenating the one-hot encoded columns with the original data
data_one_hot_encoded = pd.concat([data_one_hot, day_one_hot, opponent_one_hot], axis=1)

# Displaying the first few rows to see the one-hot encoded columns
data_one_hot_encoded.head()
#print(data_one_hot_encoded.columns)
```

```python
# Mapping for encoding the Result column
result_mapping = {'W': 2, 'D': 1, 'L': 0}

# Applying the mapping to the Result column in datasets
data_one_hot_encoded['Result_Encoded'] = data_one_hot_encoded['Result'].map(result_mapping)


# Displaying the first few rows of the training data to see the original Result column and the new Result_Encoded column
data_one_hot_encoded.head()
#print(data_one_hot_encoded.columns)
```
